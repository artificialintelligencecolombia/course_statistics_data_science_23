```python
# Create a sample dataset with 7 columns and 20 rows related to finance

np.random.seed(1911)  # for reproducibility

# Sample data generation
dates = pd.date_range('2023-01-01', periods=20, freq='B')  # 20 business days
symbols = ['AAPL', 'MSFT', 'GOOG', 'AMZN', 'FB']  # stock symbols
open_prices = np.random.uniform(100, 200, size=20).round(2)  # stock opening prices
close_prices = open_prices * np.random.uniform(0.95, 1.05, size=20).round(2)  # stock closing prices
volumes = np.random.randint(1000000, 5000000, size=20)  # trading volume
highs = np.maximum(open_prices, close_prices) * np.random.uniform(1.01, 1.05, size=20).round(2)  # daily highs
lows = np.minimum(open_prices, close_prices) * np.random.uniform(0.95, 0.99, size=20).round(2)  # daily lows

# Recreating the DataFrame with the corrected data
finance_data = pd.DataFrame({
    'Date': dates,
    'Symbol': np.random.choice(symbols, size=20),
    'Open': open_prices,
    'Close': close_prices,
    'High': highs,
    'Low': lows,
    'Volume': volumes
})

finance_data.head(20)  # Displaying the complete dataset
```
